### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold = df[df['Medal'] == 'Gold']
sports = []
for sport in gold['Sport'].unique():
  sports.append([sport,len(gold[gold['Sport'] == sport])])

sports = pd.DataFrame(sports,columns=['Sport','Gold_Medals'])
sports = sports.sort_values(by ='Gold_Medals', ascending = False)
print("********************************************************************")
print("All the sports with total number of Gold Medals they won are as follows:")
print("*********************************************************************\n")
print(sports)

print("\n*******************************************************************")
print(" 'Aquatics' sport is having most number of Gold Medals so far.")
print("*********************************************************************")
print("Top 5 Sports having most number of Gold Medals are :\n \n1] Aquatics   - 1421 \n2] Athletics  - 1215  \n3] Rowing     - 890 \n4] Gymnastics - 820 \n5] Fencing    - 552")
print("-------------------------------------------------------------------------")
print("Graph of Top 5 Sports having most number of Gold Medals is as follows:")
print("-------------------------------------------------------------------------")
sports = sports.sort_values(by ='Gold_Medals', ascending = False).head()
sports.plot(x ='Sport',y ='Gold_Medals',kind = 'bar',figsize =(12,5),color = 'blue')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)
